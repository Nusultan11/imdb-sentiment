```python
# Cell 1: project setup
from pathlib import Path
import os
import sys
import json

repo_root = Path.cwd()
if not (repo_root / "src").exists():
    if (repo_root / "imdb-sentiment" / "src").exists():
        repo_root = repo_root / "imdb-sentiment"
    elif (repo_root.parent / "src").exists():
        repo_root = repo_root.parent

os.chdir(repo_root)

src_path = repo_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

print("repo_root =", repo_root)
print("src_path =", src_path)
```

```python
import pandas as pd

from imdb_sentiment.settings import load_config
from imdb_sentiment.pipelines.model_comparison import compare_models, import_lstm_bundle
```

```python
regex_config_path = "configs/experiments/lstm_bidirectional_masked_mean_optuna_regexprep_v1.yaml"
regex_bundle_path = (
    repo_root
    / "artifacts"
    / "experiments"
    / "lstm"
    / "bidirectional_masked_mean_optuna_regexprep_v1"
    / "regex_v2_bilstm_masked_pooling_trained_bundle.zip"
)

regex_config = load_config(regex_config_path)

if regex_bundle_path.exists():
    imported = import_lstm_bundle(regex_config, regex_bundle_path)
    print(json.dumps(imported, indent=2, ensure_ascii=False))
else:
    print(f"Bundle not found: {regex_bundle_path}")
```

```python
candidate_configs = [
    "configs/experiments/tfidf_tuned_v2_final.yaml",
    "configs/experiments/lstm_baseline_v1.yaml",
    "configs/experiments/lstm_bidirectional_v1.yaml",
    "configs/experiments/lstm_bidirectional_masked_mean_v1.yaml",
    "configs/experiments/lstm_bidirectional_masked_mean_optuna_regexprep_v1.yaml",
]

candidate_configs
```

```python
report = compare_models(candidate_configs)

print(json.dumps(report["outputs"], indent=2, ensure_ascii=False))
print()
print("Winner:")
print(json.dumps(report["winner"], indent=2, ensure_ascii=False))
```

```python
comparison_dir = repo_root / "artifacts" / "reports" / "model_comparison"

results_df = pd.read_csv(comparison_dir / "all_models_test_metrics.csv")
with open(comparison_dir / "winner_summary.json", "r", encoding="utf-8") as f:
    winner_summary = json.load(f)
with open(comparison_dir / "missing_models.json", "r", encoding="utf-8") as f:
    missing_models = json.load(f)

results_df
```

```python
missing_df = pd.DataFrame(missing_models)
missing_df
```

```python
duplicate_check = results_df.copy()

metric_cols = [
    "accuracy",
    "precision",
    "recall",
    "f1",
    "preprocessing",
    "pooling",
    "bidirectional",
    "decision_threshold",
]

duplicate_groups = (
    duplicate_check
    .groupby(metric_cols, dropna=False)["model"]
    .apply(list)
    .reset_index(name="models")
)

duplicate_groups[duplicate_groups["models"].apply(len) > 1]
```

```python
print("Winner on test:")
print(f"  model: {winner_summary['winner_model']}")
print(f"  family: {winner_summary['winner_family']}")
print(f"  accuracy: {winner_summary['metrics']['accuracy']:.6f}")
print(f"  precision: {winner_summary['metrics']['precision']:.6f}")
print(f"  recall: {winner_summary['metrics']['recall']:.6f}")
print(f"  f1: {winner_summary['metrics']['f1']:.6f}")
print()
print("Runtime details:")
print(json.dumps(winner_summary["runtime_details"], indent=2, ensure_ascii=False))
```

```python
for row in results_df.to_dict(orient="records"):
    print(
        f"{row['model']}: "
        f"family={row['family']}, "
        f"accuracy={row['accuracy']:.6f}, "
        f"precision={row['precision']:.6f}, "
        f"recall={row['recall']:.6f}, "
        f"f1={row['f1']:.6f}, "
        f"preprocessing={row['preprocessing']}, "
        f"pooling={row['pooling']}, "
        f"bidirectional={row['bidirectional']}, "
        f"threshold={row['decision_threshold']}"
    )
```

```python
comparison_dir = Path("artifacts/reports/model_comparison")
comparison_dir.mkdir(parents=True, exist_ok=True)

results_df.to_csv(comparison_dir / "all_models_test_metrics.csv", index=False)
results_df.to_json(
    comparison_dir / "all_models_test_metrics.json",
    orient="records",
    indent=2,
)

with open(comparison_dir / "winner_summary.json", "w", encoding="utf-8") as f:
    json.dump(winner_summary, f, indent=2, ensure_ascii=False)

with open(comparison_dir / "missing_models.json", "w", encoding="utf-8") as f:
    json.dump(missing_df.to_dict(orient="records"), f, indent=2, ensure_ascii=False)

print("Saved to:")
print(comparison_dir / "all_models_test_metrics.csv")
print(comparison_dir / "all_models_test_metrics.json")
print(comparison_dir / "winner_summary.json")
print(comparison_dir / "missing_models.json")
```

```python
report_snapshot = {
    "results": results_df.to_dict(orient="records"),
    "winner": winner_summary,
    "missing": missing_df.to_dict(orient="records"),
}

with open(comparison_dir / "evaluation_report_snapshot.json", "w", encoding="utf-8") as f:
    json.dump(report_snapshot, f, indent=2, ensure_ascii=False)

print(comparison_dir / "evaluation_report_snapshot.json")
```
